```
import pandas as pd 

# Merge the 3 datasets from telecom 

data = pd.read_csv("/content/Telcom_dataset.csv") 
# Preview the first 5 lines of the loaded data 
data


# Read data from file 'filename.csv' ) 
data1 = pd.read_csv("/content/Telcom_dataset2.csv") 
# Preview the first 5 lines of the loaded data 
data1

# Read data from file 'filename.csv' 
data2 = pd.read_csv("/content/Telcom_dataset3.csv") 
# Preview the first 5 lines of the loaded data 
data2

M3 = pd.concat([data, data1,data2]) #merge
M3=M3.reset_index(drop=True)
M3.head(5)
```

```
#Concatinate the 3 combined sets with the call geo dataset 


M3.drop_duplicates(subset=['CELL_ID'], inplace=True) 
Geo.merge(M3, how='left', on='CELL_ID' )

MergedSet= [Geo,data, data1, data2]
M4 = pd.concat(MergedSet)

M4.head(5)
```

```
#checking for duplicates and deleting them
M4.duplicated()
M4.drop_duplicates()
```

```
#general information

M4.info()
```
